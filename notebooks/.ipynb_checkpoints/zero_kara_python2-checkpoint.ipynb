```python
import numpy as np
```

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,6,0.1)
```

↑０から６まで０．１刻みで生成

```python
y = np.sin(x)
```

```python
plt.plot(x, y)
```

```python
plt.show()
```

```python
y1= np.sin(x)
```

```python
y2= np.cos(x)
```

```python
plt.plot(x,y1, label="sin")
```

```python
plt.plot(x,y2, linestyle = "--", label="cos")
```

```python
plt.xlabel("x")
```

```python
plt.ylabel("y")
```

```python
plt.title('sin & cos')
```

```python
plt.legend()
```

```python
plt.show()
```
